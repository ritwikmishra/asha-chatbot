```python
#model file map
mfm = {}
mfm['sort_by_normalized_score_only outer.csv'] = 'Ensemble-2.1 Esc'
mfm['sort_by_normalized_score_only inner.csv'] = 'Ensemble-2.1 Esc'
mfm['sort_by_f1_score_only inner.csv'] = 'Ensemble-1.1 Ef'
mfm['sort_by_f1_score_only outer.csv'] = 'Ensemble-1.1 Ef'
mfm['sort_by_added_internal_normalized_score_only inner.csv'] = 'Ensemble-1.2 Ens'
mfm['sort_by_added_internal_normalized_score_only outer.csv'] = 'Ensemble-1.2 Ens'
mfm['sort_by_added_internal_normalized_score_bucket and f_score (group sort) inner.csv'] = 'Ensemble-1.3 Efns'
mfm['sort_by_added_internal_normalized_score_bucket and f_score (group sort) outer.csv'] = 'Ensemble-1.3 Efns'
mfm['sort_by_added_normalized_score_bucket and f_score (group sort) outer.csv'] = 'Ensemble-2.2 Efsc'
mfm['sort_by_added_normalized_score_bucket and f_score (group sort) inner.csv'] = 'Ensemble-2.2 Efsc'
mfm['ganga_inner.csv'] = 'ganga inner'
mfm['kaali_inner.csv'] = 'kaali inner'
mfm['renu_inner.csv'] = 'renu inner'
mfm['ganga_outer.csv'] = 'ganga outer'
mfm['kaali_outer.csv'] = 'kaali outer'
mfm['renu_outer.csv'] = 'renu outer'
```

```python
def get_model_name(file_name):
    global mfm
    try:
        model_name = mfm[file_name]
    except:
        model_name = '-'
    return model_name
```

```python
inner = []
outer = []
for x in sort_files:
    try:
        model = mfm[x]
        if "inner" in x:
            inner.append(x)
        else:
            outer.append(x)
    except:
        pass
```

```python
def get_cat_metrices(df1):
    categories = df1.Category.unique()
    grp = df1.groupby(["Category"])
    final_data = []
    for category in categories:
        grp_data = grp.get_group(category)
        cdict = metrices.calc_metrics(grp_data)
        cdict["category"] = category
        final_data.append(cdict)
    return final_data
```

```python
# =================== INNER AND OUTER RESULTS IN 3D =====================
%matplotlib qt5
import matplotlib.pyplot as plt

def rgb(a,b,c):
    return (a/255,b/255,c/255)

def plot3d(zll, message=''):
    fig = plt.figure(figsize=(8, 3))
    ax1 = fig.add_subplot(111, projection='3d')

    o = 0.1
    width = 0.2
    depth = 0.3
    offx = -0.25
    offy = 0.70
    bottom = 0

    xll = [[i]*8 for i in range(5)]
    yll = [[0,1,2,3,4,5,6,7]]*5
    
    cl = ['#FE0000','#FFDE00','#05BF00','#0600FF','#9500C0']

    ci = 0
    for xl,yl,zl in zip(xll,yll,zll):
        for x,y,z in zip(xl,yl,zl):
            c = (rgb(135,206,235)) if z == max(zl) else (1,1,1)
            ax1.bar3d(x-offx, y-offy, bottom, width, depth, z, shade=True, color=cl[ci], alpha=1 if z==max(zl) else o, edgecolor='black')
        ci+=1

    plt.xlim(0, 5) 
    plt.ylim(-1, 7) 

    ax1.set_zlabel('score')
    plt.title(message)
    plt.xticks([0,1,2,3,4], [r'$mAP$', r'$MRR$', r'$SR$', r'$nDCG$', r'$P@3$'])
    plt.yticks([0,1,2,3,4,5,6,7], [r'$DTP$', r'$BERT_{spc}$', r'$BERT_{cos}$', r'$\mathcal{E}_{f}$', r'$\mathcal{E}_{ns}$',r'$\mathcal{E}_{fns}$',r'$\mathcal{E}_{sc}$',r'$\mathcal{E}_{fsc}$'])
    plt.show()
    return
    
#outer
zll = [
    [18.3,21.9,10.5,23.6,24.0,23.7,23.2,24.0],
    [26.1,29.7,16.2,33.0,34.0,33.7,32.0,33.1],
    [16.4, 19.9, 8.70, 20.4, 19.4, 20.0, 19.6, 20.8],
    [27.7, 31.2, 17.4, 34.5, 35.3, 35.1, 33.6, 34.5],
    [16.4, 19.9, 8.70, 20.4, 19.4, 20.0, 19.6, 20.8]
]

plot3d(zll)

#inner
zll = [
    [41.4, 42.4, 34.0, 47.8, 49.7, 51.1, 48.4, 52.5],
    [53.4, 57.4, 49.2, 62.0, 66.2, 66.9, 64.8, 69.8],
    [35.4, 35.4, 25.3, 40.9, 40.9, 40.1, 40.1, 43.5],
    [54.0, 58.2, 50.8, 62.8, 66.9, 67.3, 65.9, 71.3],
    [35.4, 35.4, 25.3, 40.9, 40.9, 40.1, 40.1, 43.5]
]

plot3d(zll)
```

```python
cat_dicts = {}
for file_ in sort_files:
    model_name = get_model_name(file_)
    display(Markdown("## "+file_+" \[ "+model_name+" \]"))
    df1 = pd.read_csv("Data/"+file_,encoding = "ISO-8859-1")
    metrices_list = get_cat_metrices(df1)
    cat_dicts[file_] = metrices_list
    display(pd.DataFrame(metrices_list))
```

```python
def get_all_categories(cat_dicts):
    all_categories = []
    for dict_value in cat_dicts.values():
        for entry in dict_value:
            all_categories.append(entry["category"])
    all_categories = set(all_categories)
    return all_categories
```

```python
all_categories = get_all_categories(cat_dicts)

cdf = pd.DataFrame([],columns=['map', 'mrr', 'avg_sr', 'avg_ndcg', 'avg_p3','Category'])

i = 0
for category in all_categories:
#     display(Markdown("## "+category))
    cat_data = []
    for key,value in cat_dicts.items():
        for entry in value:
            if entry["category"] == category:
                temp = entry
                temp["file"] = key
                temp["Model_name"] = get_model_name(key)
                cat_data.append(temp)
    df = pd.DataFrame(cat_data)
#     display(df)
    df = df.iloc[[5,6,7,1,2,3,0,4],:]
#     zll = np.round(df[['map', 'mrr', 'avg_sr', 'avg_ndcg', 'avg_p3']].to_numpy()*100,1).transpose()
#     plot3d(zll, message=category)
    cdf.at[i,'Category'] = category
    for c in ['map', 'mrr', 'avg_sr', 'avg_ndcg', 'avg_p3']:
        cdf.at[i,c] = list(df[df[c]==df[c].max()]['Model_name'])
    i+=1

display(cdf)

```
