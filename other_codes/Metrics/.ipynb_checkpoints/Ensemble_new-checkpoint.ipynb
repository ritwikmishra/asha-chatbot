```python
def clean(hi_str):
    puncts = string.punctuation+"‟“’❝❞‚‘‛❛❜❟"
    trans_string = hi_str.maketrans("१२३४५६७८९०","1234567890")
    hi_str = hi_str.translate(trans_string)
    hi_str = re.sub(r'[a-z]+','',hi_str.lower())
    hi_str = hi_str.translate(str.maketrans(puncts,' '*len(puncts))).lower().replace('|','').replace('।','')
    hi_str = re.sub(r'\s+',' ',hi_str).strip()
    return hi_str

def clean2(hi_str):
    hi_str = hi_str.translate(str.maketrans('','',string.punctuation+"‟“’❝❞‚‘‛❛❜❟")).lower().replace('|','').replace('।','')
    hi_str = re.sub(r'\s+',' ',hi_str).strip()
    return hi_str

def lemmatize_keep_nouns(s, nlp):
    s = clean(s)
    doc = nlp(s)
    ml = []
    for sentence in doc.sentences:
        for word in sentence.words:
            if word.pos in ['NOUN','PROPN','VERB']:
                ml.append(word.lemma)
    s = ' '.join(ml)
    return s

def wfm(s1,s2): # word f measure
#     print(s1,'||',s2)
    s1 = set(lemmatize_keep_nouns(s1,nlp).split())
    s2 = set(lemmatize_keep_nouns(s2,nlp).split())
#     print(s1)
#     print(s2)
    p = (len(s1) - len(s1-s2))/(len(s2)) if len(s2)!=0 else 0
    r = (len(s1) - len(s1-s2))/(len(s1)) if len(s1)!=0 else 0
    f = ((2*p*r)/(p+r)) if p+r != 0 else 0
#     print(p,r,f)
#     input('wait')
    return f
    

def wfm_list(l1,l2): # l1 is list of suggestions at rank 1, l2 is user query. Length of both are same.
    ml = []
    assert len(l1) == len(l2)
    for s1,s2 in zip(l1,l2): # take one by one
        ml.append(wfm(s1,s2))
#     print(ml)
#     input('here')
    return ml
        
```

```python
topk = 3
df1 = pd.read_csv('/home/ritwik/git/AshaQnA/best results/Outside_questions_BERT_spc_results_695_epoch_3.csv') 
df2 = pd.read_csv('/home/ritwik/git/AshaQnA/best results/Outside_questions_DTP_results_added_words.csv')
df4 = pd.read_csv('/home/ritwik/git/AshaQnA/best results/Outside_questions_BERT_cos_results_added_words_mpnet.csv')

# df1 = pd.read_csv('/home/ritwik/git/AshaQnA/best results/top5/Outside_questions_BERT_cos_results_added_words_mpnet_top5.csv')
# df2 = pd.read_csv('/home/ritwik/git/AshaQnA/best results/top5/Outside_questions_BERT_spc_results_695_epoch_3_top5.csv')
# df4 = pd.read_csv('/home/ritwik/git/AshaQnA/best results/top5/Outside_questions_DTP_results_top5.csv')
# topk = 5

cdf = [df1,df2,df4]

print([len(x) for x in cdf])

df = pd.concat(cdf)
df.loc[0,:].reset_index(drop=True)
```

```python
# --------- E_f wfm scores -----------
for i in range(topk):
    if 'wfm_'+str(i+1) not in df.columns:
        df.insert(list(df.columns).index('final_'+str(i+1)),'wfm_'+str(i+1),'')
#         df['wfm_'+str(i+1)] = 0.0
#         df['wfm_'+str(i+1)] = df['wfm_'+str(i+1)].astype(float)

for i in tqdm(range(len(df1)),desc='wfm'):
    tdf = df.loc[i,:].reset_index(drop=True)
    for j in range(topk):
        df.at[i,'wfm_'+str(j+1)] = list(wfm_list(tdf['OPTION_'+str(j+1)], tdf['User_query']))
#     display(df.loc[i,:])    
#     input()
display(df.loc[0,:])
```

```python
# --------- E_ns scores -----------

for i in range(topk):
    if 'ns_'+str(i+1) not in df.columns:
        df.insert(list(df.columns).index('final_'+str(i+1)),'ns_'+str(i+1),'')

for i in tqdm(range(len(df1)),desc='wfm'):
    tdf = df.loc[i,:].reset_index(drop=True)
    mxl = []
    for j in range(len(tdf)):
        tl = []
        for k in range(topk):
            tl.append(tdf.loc[j,'SCORE_'+str(k+1)])
        mxl.append(max(tl))
    for j in range(topk):
        df.at[i,'ns_'+str(j+1)] = df.loc[i,'SCORE_'+str(j+1)]/mxl
display(df.loc[0,:])
```

```python
# --------- E_sus -----------

same_score = pd.DataFrame([],columns=[x for x in range(len(cdf))]+['opt','row']) # opt represents suggestion rank, row represents user query number
adf = []

count = 0
for i in tqdm(range(len(df1))):
    tdf = df.loc[i,:].reset_index(drop=True)
    for optnum in range(topk):
        opts = list(tdf['OPTION_'+str(optnum+1)].unique())
        for o in opts:
            #ttdf = tdf[tdf['OPTION_'+str(optnum+1)]==o]
            ttdf = tdf[(tdf['OPTION_'+str(optnum+1)]==o) & (tdf['final_'+str(optnum+1)].apply(lambda d: d in [1,2]))]
            if len(ttdf) > 1:
                same_score.at[count,ttdf.index] = ttdf['SCORE_'+str(optnum+1)]
                same_score.at[count,'opt'] = optnum+1
                same_score.at[count,'row'] = i
                count+=1
display(same_score)
same_score = same_score[[x for x in range(len(cdf))]] # removing "opt" and "row" columns
for c in same_score.columns:
    sl = []
    tdf = same_score[same_score[c].notnull()]
    tdf = tdf.fillna(0)
    for _, row in tdf.iterrows():
        sl.append(max(row)/row[c])
    print('Adjustment factor of',c,'is:',sum(sl)/len(sl))
    adf.append(sum(sl)/len(sl))

for i in range(topk):
    if 'sc_'+str(i+1) not in df.columns:
        df.insert(list(df.columns).index('final_'+str(i+1)),'sc_'+str(i+1),'')

for i in tqdm(range(len(df1)),desc='wfm'):
    tdf = df.loc[i,:].reset_index(drop=True)
    for j in range(topk):
        df.at[i,'sc_'+str(j+1)] = [min(x,1.0) for x in list(tdf['SCORE_'+str(j+1)]*adf)]
    #display(df.loc[i,:])    
#     print(df.dtypes)
#     input('wait')
display(df.loc[0,:])

# old
# Adjustment factor of 0 is: 1.1639364089550563
# Adjustment factor of 1 is: 1.6852641610531967
# Adjustment factor of 2 is: 1.0010583234164507

# new
# Adjustment factor of 0 is: 1.0
# Adjustment factor of 1 is: 2.1803420976692607
# Adjustment factor of 2 is: 1.1325624354513701
    
    # new v2
    # Adjustment factor of 0 is: 1.005626816574814
    # Adjustment factor of 1 is: 1.9931843556536775
    # Adjustment factor of 2 is: 1.0555843010504093
    
    # without cos
    # Adjustment factor of 0 is: 1.0
    # Adjustment factor of 1 is: 2.084327003740152

# new top 5
# Adjustment factor of 0 is: 1.1325624354513701
# Adjustment factor of 1 is: 1.0
# Adjustment factor of 2 is: 2.1803420976692607

# new without COS
# Adjustment factor of 0 is: 1.0
# Adjustment factor of 1 is: 2.4360442789603365
```

```python
df_e = pd.DataFrame([], columns=df1.columns)
for i in tqdm(range(len(df1))):
    tdf = df.loc[i,:].reset_index(drop=True)
    df_e.at[i,:] = tdf.loc[0,:]
    count = 0
    score_cols = ['SCORE'
#                   ,'wfm','sc','wfm_sc','wfm_only'
#                   ,'ns'
                 ]
    data_cols = ['OPTION','ANSWER','final']
    target_score = 'SCORE'
    stacked_a = pd.DataFrame([],columns=data_cols+score_cols)
    for j in range(len(tdf)):
        for k in range(topk):
            data = []
            for col in stacked_a.columns:
                data.append(tdf.loc[j,col+'_'+str(k+1)])
            stacked_a.at[count,:] = data
            count+=1
    stacked_a_merged = pd.DataFrame([], columns=stacked_a.columns)
    # -------- Add the scores of repeating suggestions --------
    count = 0
    for opt in stacked_a['OPTION'].unique():
        option_batch = stacked_a[stacked_a['OPTION'] == opt].reset_index(drop=True)
        for col in data_cols:
            stacked_a_merged.at[count,col] = option_batch.loc[0,col]
        for col in score_cols:
            stacked_a_merged.at[count,col] = sum(option_batch[col])
        count+=1
    stacked_a_merged = stacked_a_merged.sort_values(target_score, ascending=False).reset_index(drop=True)
    # ---------------------------------------------------------
    
    # ------ Do not add the scores of repeating suggestions, but make sure the top-k are all unique ---------
#     stacked_a = stacked_a.sort_values(target_score, ascending=False).reset_index(drop=True)
#     count = 0
#     for j in range(len(stacked_a)):
#         if stacked_a.loc[j,'OPTION'] not in list(stacked_a_merged['OPTION']):
#             stacked_a_merged.at[count,:] = stacked_a.loc[j,:]
#             count+=1
    # -------------------------------------------------------------------------------------------------------
    if len(stacked_a_merged) < len(stacked_a) and False: # for showing intermediate steps
        display(tdf)
        display(stacked_a)
        print(list(df.loc[i,'User_query'])[0])
        display(stacked_a_merged)
        input()
    for j in range(topk):
        df_e.at[i,'OPTION_'+str(j+1)] = stacked_a_merged.loc[j,'OPTION']
        df_e.at[i,'ANSWER_'+str(j+1)] = stacked_a_merged.loc[j,'ANSWER']
        df_e.at[i,'SCORE_'+str(j+1)] = stacked_a_merged.loc[j,target_score]
        df_e.at[i,'final_'+str(j+1)] = stacked_a_merged.loc[j,'final']
    df_e.at[i,'Model_name'] = 'Ensemble '+target_score

df_e.to_csv('/home/ritwik/git/AshaQnA/other_codes/Metrics/Data/E_49.csv',index=None)
```
